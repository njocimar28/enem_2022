```python
# Contagem das presenças para cada classe
contagem_distintos_cn = dados_enem_2022['TP_PRESENCA_CN'].value_counts()
contagem_distintos_ch = dados_enem_2022['TP_PRESENCA_CH'].value_counts()
contagem_distintos_lc = dados_enem_2022['TP_PRESENCA_LC'].value_counts()
contagem_distintos_mt = dados_enem_2022['TP_PRESENCA_MT'].value_counts()

# Definir a largura das barras
bar_width = 0.2

# Definir posições para as barras
pos_lc = np.arange(len(contagem_distintos_cn))
pos_ch = pos_lc + bar_width
pos_cn = pos_ch + bar_width
pos_mt = pos_cn + bar_width

# Criar um gráfico de barras
plt.figure(figsize=(8, 4))

plt.bar(pos_cn, contagem_distintos_cn.values, bar_width, label='Ciências da Natureza', alpha=0.7)
plt.bar(pos_ch, contagem_distintos_ch.values, bar_width, label='Ciências Humanas', alpha=0.7)
plt.bar(pos_lc, contagem_distintos_lc.values, bar_width, label='Linguagens e Códigos', alpha=0.7)
plt.bar(pos_mt, contagem_distintos_mt.values, bar_width, label='Matemática', alpha=0.7)

plt.xlabel('Presença')
plt.ylabel('Contagem')
plt.title('Presenças por Classe de Prova')
plt.legend()
plt.xticks(pos_cn + bar_width * 1.5, ['Presente', 'Faltou', 'Eliminado'])

plt.tight_layout()
plt.show()
```

```python
# Contagem de inscrições
numero_de_inscricoes = dados_enem_2022.shape[0]
print("Número de incrições:", numero_de_inscricoes)

# Contagem de inscrições
dados_enem_2022_sem_zeros = dados_enem_2022.query("nota_media_final != 0.00")
numero_de_inscricoes_sem_zeros = dados_enem_2022_sem_zeros.shape[0]
print("Número de incrições sem  media zero:", numero_de_inscricoes_sem_zeros)

# representação percentual de incritos que não tiveram nota zero
percentual_sem_zeros = ((numero_de_inscricoes_sem_zeros / numero_de_inscricoes) * 100)
print("Percentual de inscritos sem media zero:", percentual_sem_zeros)
```

```python
# Dados
categorias = ['Sem Média Zero', 'Com Média Zero']
valores = [percentual_sem_zeros, 100 - percentual_sem_zeros]

# Cores das fatias
cores = ['green', 'gray']

# Crie uma figura e um eixo
plt.figure(figsize=(4, 3))

# Crie o gráfico de pizza
plt.pie(valores, labels=categorias, colors=cores, autopct='%1.1f%%', startangle=140)

# Defina o título do gráfico
plt.title("Percentual de Inscritos sem Média Zero")

# Exiba o gráfico
plt.axis('equal')  # Mantém o gráfico de pizza com proporções iguais
plt.show()
```

```python
# Agrupar os dados por renda e calcular a média de notas por prova
nota_final_por_renda = pd.DataFrame(dados_enem_2022_sem_zeros.groupby('classe_social')['nota_media_final'].mean().round(2))

nota_final_por_renda.head(5)
```

```python
# Criar figura e eixos
fig, ax = plt.subplots(figsize=(6, 4))

# Ordenar as classes sociais pela média de notas
nota_final_por_renda_sorted = nota_final_por_renda.sort_values(by='nota_media_final', ascending=False)

# Plotar as médias das notas em MT por renda
bars = nota_final_por_renda_sorted.plot(kind='bar', ax=ax, color='skyblue')

# Adicionar os rótulos de dados nas barras
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=8)

# Configurar os rótulos e títulos
plt.title('Média de Notas Finais por Classe Social')
plt.xlabel('Classe Social (Renda Mensal)')
plt.ylabel('Média de Notas Finais')
plt.xticks(rotation=0, ha='right')
ax.legend().set_visible(False)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
